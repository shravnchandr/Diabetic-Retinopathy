```python
import numpy
import pandas
from tqdm import tqdm

import cv2
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, cohen_kappa_score

model = XGBClassifier()
```

```python
df = pandas.read_csv("../input/dr-augmented/dr.csv")
folder = "../input/dr-augmented/dr/upload/"

images = []
labels = []
for idx in tqdm(range(len(df))):
    image = cv2.imread(folder+df.iloc[idx]['id_code']+'.png')
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    images.append(cv2.resize(image, (128, 128)))
    labels.append(df.iloc[idx]['diagnosis'])
    
images = numpy.array(images)
images = images.reshape(images.shape[0],images.shape[1]*images.shape[2]*images.shape[3])
labels = numpy.array([0 if value == 0 else 1 for value in labels])

x_train, x_test, y_train, y_test = train_test_split(images, labels, test_size=0.2,
                                                   random_state=42, shuffle=True, stratify=labels)
```

```python
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
y_pred = [round(value) for value in y_pred]

print(accuracy_score(y_test, y_pred))
print(cohen_kappa_score(y_test, y_pred))
```
